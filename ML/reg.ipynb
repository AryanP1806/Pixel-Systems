```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("csv/Salary_dataset.csv")
dataset.drop(columns=["Unnamed: 0"], inplace=True)
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
x = dataset[["YearsExperience"]]
y = dataset["Salary"]
```

```python
# plt.figure(figsize=(10,5))
sns.scatterplot(x="YearsExperience", y="Salary", data=dataset)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
pred = lr.predict(x)
lr.predict([[6]])
```

```python
lr.score(x_test,y_test)
```

```python
y = lr.coef_ * 6 + lr.intercept_
y
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x="YearsExperience", y="Salary", data=dataset)
plt.plot(dataset["YearsExperience"],pred,c="red")
plt.legend(["Salary", "Prediction"] )
# plt.savefig("pred.jpg")
plt.show()
```
