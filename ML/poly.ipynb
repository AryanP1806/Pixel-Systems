```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv("csv/Ice_cream selling data.csv")
dataset.head(3)
```

```python
dataset.corr()
```

```python
plt.scatter(dataset["Temperature (°C)"],dataset["Ice Cream Sales (units)"])
plt.xlabel("Temp")
plt.ylabel("Sales")
```

```python
x = dataset[["Temperature (°C)"]]
y = dataset["Ice Cream Sales (units)"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=3)
pf.fit(x)
x = pf.transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=75)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test,y_test)
```

```python
pred = lr.predict(x)

plt.scatter(dataset["Temperature (°C)"],dataset["Ice Cream Sales (units)"])
plt.plot(dataset["Temperature (°C)"],pred,c="red")
plt.xlabel("Temp")
plt.ylabel("Sales")
```
